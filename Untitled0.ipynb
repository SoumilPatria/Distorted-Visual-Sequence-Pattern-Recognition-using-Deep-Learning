## **🧠 Project Overview: Distorted Text Recognition (OCR)**

This code implements a Convolutional Recurrent Neural Network (CRNN) designed to read and transcribe distorted text from images. It mimics how an AI "reads" by combining visual feature extraction with sequential language modeling.

**How the Pipeline Works:**

**1) Data Preparation:** The code automatically detects your dataset folders, converts images to grayscale, and encodes text labels into numeric sequences that the neural network can process.

**2) The CRNN Architecture:**

**CNN Layers (The "Eyes"):** Uses multiple layers of convolutions to scan the images and extract visual patterns like edges, curves, and character shapes.

**Bi-LSTM Layers (The "Reader"):** Since text is sequential, a Bidirectional Long Short-Term Memory (LSTM) network processes the extracted visual features from left-to-right (and right-to-left) to understand the context of the characters.

**3) CTC Loss (The "Grader"):** Because the length of the text in the image may not align perfectly with the network's output steps, the Connectionist Temporal Classification (CTC) loss function is used. It allows the model to learn the correct character sequence without needing to know the exact pixel-perfect alignment of each letter.

**4) Training & Inference:** The model undergoes several training rounds where it calculates the "error" between its guess and the actual label, then adjusts its internal weights. Finally, it uses a Greedy Decoder to turn the network's raw probability outputs into a human-readable text string for the final submission.

**5) Evaluation**
The performance is measured using Character Error Rate (CER), which utilizes Levenshtein Distance to calculate the number of insertions, deletions, and substitutions required to match the predicted text to the actual target. A lower CER score indicates a higher level of accuracy and better model performance.

```python
# ==============================================================================
# 1. DEPENDENCIES & ENVIRONMENT SETUP
# ==============================================================================
# Install Levenshtein distance package required for accuracy evaluation
!pip install Levenshtein

import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import pandas as pd
import numpy as np
from PIL import Image
import Levenshtein

# ==============================================================================
# 2. CONFIGURATION & RULES
# ==============================================================================
# Settings for training the neural network
images_per_batch = 32         # How many images the model looks at before updating itself
learning_speed = 0.0001       # How aggressively the model adjusts its guesses
total_training_rounds = 30    # Total times the model will review the entire dataset
image_height = 64             # Standardized height for all incoming images
image_width = 256             # Standardized width for all incoming images

# Automatically use the GPU for faster math if it's available
compute_device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- SMART FOLDER AUTO-DETECTION ---
# Locates the data no matter where this notebook is running
possible_locations = ["", "cig_ps/", "cig_data/"]
active_folder = ""

for folder in possible_locations:
    if os.path.exists(os.path.join(folder, "train-labels.csv")):
        active_folder = folder
        break

# Assign final paths based on what the script found
training_labels_file = os.path.join(active_folder, "train-labels.csv")
training_images_folder = os.path.join(active_folder, "train_images")
testing_images_folder = os.path.join(active_folder, "test_images")

# The alphabet the model is allowed to use
allowed_characters = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Dictionaries to translate characters to numbers (and vice-versa) for the neural network
character_to_number = {char: index + 1 for index, char in enumerate(allowed_characters)}
number_to_character = {index: char for char, index in character_to_number.items()}

# Total categories = 36 characters + 1 mandatory "blank" space for the CTC algorithm
total_unique_characters = len(allowed_characters) + 1

print(f"Hardware Check: Using -> {compute_device}")
print(f"Data Location Found: '{active_folder if active_folder else './'}'")

# ==============================================================================
# 3. PREPARING THE DATA
# ==============================================================================
class TextImageDataset(Dataset):
    """
    Our custom loader that grabs an image, turns it black and white,
    and translates the text answer into numbers the model can understand.
    """
    def __init__(self, data_table, image_folder, image_filters=None, is_testing_mode=False, image_column='image', text_column='text'):
        self.data_table = data_table
        self.image_folder = image_folder
        self.image_filters = image_filters
        self.is_testing_mode = is_testing_mode
        self.image_column = image_column
        self.text_column = text_column

    def __len__(self):
        return len(self.data_table)

    def __getitem__(self, row_index):
        filename = self.data_table.iloc[row_index][self.image_column]
        full_path = os.path.join(self.image_folder, filename)

        # Open image and strip out color (grayscale)
        raw_image = Image.open(full_path).convert('L')

        # Apply visual tweaks (resizing, brightness adjustments, etc.)
        if self.image_filters:
            raw_image = self.image_filters(raw_image)

        # If we are testing, there is no answer key to return
        if self.is_testing_mode:
            return raw_image, filename

        # For training, grab the right answer and convert it into a number list
        correct_text = str(self.data_table.iloc[row_index][self.text_column]).upper()
        number_sequence = [character_to_number[char] for char in correct_text if char in character_to_number]

        return raw_image, torch.tensor(number_sequence, dtype=torch.long)

def package_batch_data(batch_items):
    """
    Organizes a group of images and their corresponding text answers.
    Because text lengths vary, we flatten the answers into one long list for the loss function.
    """
    images, text_answers = zip(*batch_items)
    stacked_images = torch.stack(images, 0)

    # Record how long each answer is before mashing them together
    answer_lengths = torch.tensor([len(answer) for answer in text_answers], dtype=torch.long)
    flattened_answers = torch.cat(text_answers)

    return stacked_images, flattened_answers, answer_lengths

# Data Augmentation: Randomly tweak training images so the model doesn't just memorize them
training_filters = transforms.Compose([
    transforms.Resize((image_height, image_width)),
    transforms.ColorJitter(brightness=0.2, contrast=0.2), # Add slight lighting differences
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,)) # Standardize pixel math
])

# Clean, predictable processing for testing
testing_filters = transforms.Compose([
    transforms.Resize((image_height, image_width)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# ==============================================================================
# 4. BUILDING THE NEURAL NETWORK
# ==============================================================================
class ImageToTextNetwork(nn.Module):
    """
    Combines a Convolutional Network (to see shapes/edges) with a
    Recurrent Network (to read left-to-right).
    """
    def __init__(self, output_options, memory_size=256):
        super(ImageToTextNetwork, self).__init__()

        # 1. The "Eyes": Extracts visual patterns from the image
        self.visual_extractor = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1), nn.ReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1), nn.ReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1), nn.BatchNorm2d(256), nn.ReLU(),
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1), nn.ReLU(),
            nn.MaxPool2d((2, 1), (2, 1)), # Shrink height but keep width to read left-to-right

            nn.Conv2d(256, 512, kernel_size=3, stride=1, padding=1), nn.BatchNorm2d(512), nn.ReLU(),
            nn.MaxPool2d((2, 1), (2, 1)),

            nn.Conv2d(512, 512, kernel_size=2, stride=1, padding=0), nn.BatchNorm2d(512), nn.ReLU()
        )

        # 2. The "Bridge": Prepares the visuals for reading
        self.bridge_layer = nn.Linear(512 * 3, memory_size)

        # 3. The "Reader": Scans the visual features forward and backward
        self.sequence_reader = nn.Sequential(
            nn.LSTM(memory_size, memory_size, bidirectional=True, num_layers=2, batch_first=False, dropout=0.3)
        )

        # 4. The "Speaker": Turns the reader's thoughts into character predictions
        self.final_translator = nn.Linear(memory_size * 2, output_options)

    def forward(self, incoming_image):
        # Look at the image
        visual_features = self.visual_extractor(incoming_image)

        # Reorganize the data structure for reading
        batch, channels, height, width = visual_features.size()
        visual_features = visual_features.view(batch, channels * height, width)
        visual_features = visual_features.permute(2, 0, 1)

        # Read the sequence
        bridged_features = self.bridge_layer(visual_features)
        reader_thoughts, _ = self.sequence_reader(bridged_features)

        # Make the final guesses
        character_predictions = self.final_translator(reader_thoughts)
        return character_predictions

# ==============================================================================
# 5. TRANSLATION UTILITIES
# ==============================================================================
def translate_math_to_text(network_output):
    """
    Turns the raw probability numbers back into human words.
    Cleans up duplicate characters and ignores the blank spaces.
    """
    # Find the character with the highest score at each step
    best_guesses = torch.argmax(network_output, dim=-1).permute(1, 0)
    final_text_results = []

    for row in best_guesses:
        built_word = []
        previous_character = None

        for number in row:
            number = number.item()
            # If it's not a blank space (0) and not a stutter/duplicate of the last letter
            if number != 0 and number != previous_character:
                built_word.append(number_to_character[number])
            previous_character = number

        final_text_results.append("".join(built_word))

    return final_text_results

# ==============================================================================
# 6. THE LEARNING PROCESS
# ==============================================================================
def run_training_round(model, data_provider, loss_calculator, optimizer_tool):
    """
    Runs one full pass over all the training images, calculating mistakes
    and updating the model's brain to do better next time.
    """
    model.train()
    total_mistakes = 0

    for batch_images, batch_answers, answer_lengths in data_provider:
        batch_images = batch_images.to(compute_device)
        batch_answers = batch_answers.to(compute_device)

        optimizer_tool.zero_grad() # Wipe the slate clean for this new batch
        predictions = model(batch_images)

        # Calculate how many predictions the model made per image
        prediction_lengths = torch.full(size=(predictions.size(1),), fill_value=predictions.size(0), dtype=torch.long).to(compute_device)

        # Grade the model's work
        mistake_score = loss_calculator(predictions.log_softmax(2), batch_answers, prediction_lengths, answer_lengths)
        mistake_score.backward() # Trace back through the math to find what went wrong

        # Put a speed limit on changes so the model doesn't overcorrect and break
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)

        optimizer_tool.step() # Apply the updates

        total_mistakes += mistake_score.item()

    return total_mistakes / len(data_provider)

# ==============================================================================
# 7. MAIN SCRIPT EXECUTION
# ==============================================================================
if __name__ == "__main__":
    if os.path.exists(training_labels_file):

        # Load the answer keys
        training_data_table = pd.read_csv(training_labels_file)

        # Find all the test images
        testing_filenames = [f for f in os.listdir(testing_images_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        testing_data_table = pd.DataFrame({'image': testing_filenames})

        print(f"Loaded {len(training_data_table)} training images and found {len(testing_data_table)} images to test on.")

        # Set up the data delivery systems
        training_provider = TextImageDataset(training_data_table, training_images_folder, image_filters=training_filters)
        training_loader = DataLoader(training_provider, batch_size=images_per_batch, shuffle=True, collate_fn=package_batch_data)

        testing_provider = TextImageDataset(testing_data_table, testing_images_folder, image_filters=testing_filters, is_testing_mode=True)
        testing_loader = DataLoader(testing_provider, batch_size=images_per_batch, shuffle=False)

        # Build the engine
        ocr_model = ImageToTextNetwork(output_options=total_unique_characters).to(compute_device)
        mistake_calculator = nn.CTCLoss(blank=0, zero_infinity=True)
        model_optimizer = optim.RMSprop(ocr_model.parameters(), lr=learning_speed)

        # Start the education process
        print("Starting the learning process...")
        for round_number in range(1, total_training_rounds + 1):
            average_mistakes = run_training_round(ocr_model, training_loader, mistake_calculator, model_optimizer)
            print(f"Round [{round_number}/{total_training_rounds}] -> Average Error Score: {average_mistakes:.4f}")

        # ==============================================================================
        # 8. FINAL EXAM & SUBMISSION
        # ==============================================================================
        ocr_model.eval() # Tell the model to stop learning and start testing
        final_image_names = []
        final_text_predictions = []

        print("Reading the test dataset...")
        with torch.no_grad(): # Save memory by turning off the learning math
            for images, filenames in testing_loader:
                images = images.to(compute_device)
                raw_predictions = ocr_model(images)
                human_readable_text = translate_math_to_text(raw_predictions)

                final_image_names.extend(filenames)
                final_text_predictions.extend(human_readable_text)

        # Package the answers exactly as required by the challenge
        submission_table = pd.DataFrame({'image': final_image_names, 'prediction': final_text_predictions})
        final_filename = "submission_Soumil_Patria_24115147.csv"
        submission_table.to_csv(final_filename, index=False)

        print(f"Success! Your final answers have been saved to: {final_filename}")
        print("Check your file sidebar to download the CSV.")

    else:
        print(f"Error: Couldn't find '{training_labels_file}'. Make sure your files are extracted correctly.")
```
